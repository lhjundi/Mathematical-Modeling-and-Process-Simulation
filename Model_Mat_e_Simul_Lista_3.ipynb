```python
# Lista 3 - Exercício 1
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

# Parâmetros
n=100
x_final=2
dx=x_final/n
delta=0
x=np.empty([n])
y=np.empty([n])
u=np.empty([n])
y2=np.empty([n])

# Condição inicial
x[0]=0
y[0]=1
u[0]=0

# Condições seguintes
for i in range(1,n,1):

# Derivadas
    dydx=u[i-1]
    dudx=y[i-1]+math.exp(x[i-1])

# Atualização
    x[i]=x[i-1]+dx
    y[i]=y[i-1]+dydx*dx
    u[i]=u[i-1]+dudx*dx

# Analítica
for i in range(0,n):
    y2[i]=1/4*math.exp(x[i])+3/4*math.exp(-x[i])+(
        1/2*x[i]*math.exp(x[i]))+delta
plt.title('Lista 3 - Exercício 1')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y, color='red', label='Numérica') #red
plt.plot(x, y2, color='blue', label='Analítica') #blue
plt.legend(loc='best')
plt.show()
```

```python
# Lista 3 - Exercício 2
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from pylab import*

# Parâmetros
n=100
x_final=math.pi
dx=x_final/n
delta=0
x=np.empty([n])
y=np.empty([n])
u=np.empty([n])
y2=np.empty([n])

# Condição inicial
x[0]=0
y[0]=0
u[0]=1

# Condições seguintes
for i in range(1,n,1):

# Derivadas
    dydx=u[i-1]
    dudx=-16*y[i-1]

# Atualização
    x[i]=x[i-1]+dx
    y[i]=y[i-1]+dydx*dx
    u[i]=u[i-1]+dudx*dx

# Analítica
for i in range(0,n):
    y2[i]=(math.sin(4*x[i])*1/4)+delta
plt.title('Lista 3 - Exercicio 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y, color='red', label='numérico') #red
plt.plot(x, y2, color='blue', label='analítico') #blue
plt.legend(loc='best')
plt.show()
```
