d) Desenvolva um programa que, aplicando o método da bisecção, resolva as seguintes questões:

d1) Deve ser consstruída uma peça, com o formado de um arco de circunferência, de comprimento 8,0 m. Essa peça será fixada em base de comprimento igual a 6,0 metros. Determine a altura da peça.
 mmsp lista 1 d1.png


```python
#d1

def  bisseccao(f, a, b, tol=1e-9, Nmax = 50):
     n=1
     while n<Nmax:
       c= (b+a)/2
       if f(c)==0 or ((b-a)/2)<tol:
         print('raíz = ', c)
         print('erro = ', f(c))
         if f(c)<=1e-2:
             return c
         elif sign(f(c)*f(a))>0:
             a = c
         else:
             b = c
             n +=1
             print('Raízes não encontradas')
```

```python
def f(R):
     return sin(8/R) - 6/R

x=linspace(2, 15, 1000)
plot(x, f(x))
xlabel('c')
ylabel('f')
title("exercício d4")
grid()
r = bisseccao(f, a=4, b=8)

plot(r, 0, 'r*')
th = arctan(6/r)
h = r*(1-cos(th))
print('a peça tem uma altura em mestros de: ', h)
```
