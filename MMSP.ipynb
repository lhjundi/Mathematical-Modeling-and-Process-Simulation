```python
import numpy as np
import math
import matplotlib.pyplot as plt
### Parâmetros ###
x_fim=1.
N=20
dx=x_fim/float(N)
x=np.empty([N+1])
y=np.empty([N+1])
y2=np.empty([N+1])
### Condição inicial ###
x[0]=0.
y[0]=0.
y2[0]=0.
### Condições seguintes ###
for i in range(1,N+1,1):
### Derivada ###
    dydx=x[i-1]+y[i-1]
### Atualização ###
    x[i]=x[i-1]+dx
    y[i]=y[i-1]+dydx*dx
    y2[i]=math.exp(x[i])-x[i]-1 #analítica
### Resultados ###
plt.plot(x,y,'r',label='numérico') #red
plt.plot(x,y2,'g',label='analítico') #green
plt.legend(loc='best')
plt.xlabel('x')
plt.grid()
plt.show
```
